### Нефтедобыча в промышленности

```python
import pandas as pd
```

```python
df = pd.read_excel('Объемы промышленного производства по видам экономической деятельности по РК (1990-2023).xlsx')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame setup (replace with actual data import from Excel)
# df = pd.read_excel('path_to_your_file.xlsx', sheet_name='YourSheetName')

# Calculating the percentage of each year's value in the first row relative to the second row
percentage_growth = (df.iloc[0] / df.iloc[1]) * 100

# Extracting the years from the columns
years = df.columns.astype(int)

# Plotting the percentage growth over the years
plt.figure(figsize=(12, 6))
plt.plot(years, percentage_growth, marker='o', color='b', linestyle='-')
plt.xlabel('Год')
plt.ylabel('Процент нефтедобычи в промышленности')
plt.title('Динамичный график процента нефтедобычи в промышленности')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('Среднемесячная заработная плата по регионам Республики Казахстан (2010-2024гг.).xlsx')
df.rename(columns={'Регионы': 'Region'}, inplace=True)
df.set_index('Region', inplace=True)
df.columns = df.columns.astype(str)
df = df.apply(pd.to_numeric, errors='coerce')
plt.figure(figsize=(12, 8))
for region in df.index:
    plt.plot(df.columns, df.loc[region], label=region)

plt.xlabel('Год')
plt.ylabel('Рост заработной платы в Регионах (2010-2024)')
plt.title('Рост заработной платы в Регионах (2010-2024)')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), title="Region")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html, Input, Output

df = pd.read_excel('Среднемесячная заработная плата по регионам Республики Казахстан (2010-2024гг.).xlsx')

df.rename(columns={'Регионы': 'Region'}, inplace=True)
df.set_index('Region', inplace=True)
df.columns = df.columns.astype(str)
app = Dash(__name__)

app.layout = html.Div([
    html.H1("Среднемесячная заработная плата по регионам Казахстана (2010-2024)"),
    
    html.Label("Выберите регион:"),
    dcc.Dropdown(
        id='region-dropdown',
        options=[{'label': region, 'value': region} for region in df.index],
        value='Республика Казахстан'  # Значение по умолчанию
    ),
    
    dcc.Graph(id='wage-graph'),
    
    html.Label("Выберите диапазон лет:"),
    dcc.RangeSlider(
        id='year-slider',
        min=int(df.columns[0]),
        max=int(df.columns[-1]),
        step=1,
        value=[int(df.columns[0]), int(df.columns[-1])],
        marks={int(year): str(year) for year in df.columns}
    )
])

@app.callback(
    Output('wage-graph', 'figure'),
    [Input('region-dropdown', 'value'),
     Input('year-slider', 'value')]
)
def update_graph(selected_region, selected_years):
    # Фильтрация данных по выбранному региону и диапазону лет
    filtered_df = df.loc[selected_region, str(selected_years[0]):str(selected_years[1])]
    
    # Построение графика
    fig = px.line(
        x=filtered_df.index,
        y=filtered_df.values,
        title=f'Среднемесячная заработная плата в регионе {selected_region}',
        labels={'x': 'Год', 'y': 'Среднемесячная зарплата (тенге)'}
    )
    fig.update_layout(xaxis_title='Год', yaxis_title='Среднемесячная зарплата (тенге)')
    return fig

# Запуск приложения
if __name__ == '__main__':
    app.run_server(debug=True)
```
